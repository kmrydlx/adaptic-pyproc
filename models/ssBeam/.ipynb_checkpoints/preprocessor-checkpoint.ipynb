# ADAPTIC PyProc preprocessor - simply supported beam

This scripts generates an input file _.dat_ for an analysis of a simply supported beam in __ADAPTIC__.

Refer to _postprocessor.ipynb_ , located in the same folder, to learn about the analysis results.

## Problem specification

A 6 m long simply supported beam with an inverted T-section is subjected to a three-point bending test. A quasi-static analysis is performed, meaning that dynamic analysis with slow load application is emplyed, to obtain better convergence.

## Input

Use the code block below to define basic analysis options.

```python
import numpy as np
import pandas as pd
from datetime import datetime as day
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import sys
sys.path.insert(1, '../../libraries')
import utils
import materials as mat
import sections as sect
```

```python
project_name = 'ssBeam'
```

```python
# starting element name numbering
startCBP = 1
startCNC = 20001
startBND = 30001
startJEL = 40001
startLNK = 50001
startCNC = 60001
startCBM = 80001
```

```python
save = True # generates *.dat and .pkl files
```

## Initialisation

Define function `add` that appends to `all_lines` variable, which stores each line of text to be printed in _*.dat_ file.

```python
def add(line): return np.vstack((all_lines, line))
```

## Define materials

Below con1 as well as stl1 materials are generated.

```python
# Concrete
concr = mat.con1('con', fc1=50, length=250)
concr.data_frame()
```

```python
# Steel
reinf = mat.stl1('reinf', E1 = 205000, fy = 500, fu = 593, epsilon_u = 0.109, softening=True, E3=5)
reinf.data_frame()
```

```python
material_lst = [concr,reinf]
```

## Define sections
### rcts sections

```python
reinforcement=[[7, 25, 700], [2, 16, 50]] # [[number, diameter, height]]
beam_section1 = sect.rcts('B1', reinf.ID, concr.ID, concr.ID, 190, 560, 650, 340, 30, 12, reinforcement)
section_rcts_lst = [beam_section1]
print('The distance to the bottom fibre from the centroid: ', -beam_section1.centr_top,' mm')
beam_section1.plotting(title='Section B1',reverse=True)
```

## Groups

```python
sections_lst = [section_rcts_lst]
sec_flat = [i for sub_l in sections_lst for i in sub_l]
sec_names = np.array([i.ID for sub_l in sections_lst for i in sub_l])
monitor_pts = 200
```

### cbp2 group

```python
cbp_lst = [['g'+i,i,monitor_pts] for i in sec_names]
```

### lnk2 group

```python
x = 'rigid'
y = 'rigid'
rz = 'rigid'
lnk_lst=[utils.param2adap('glnkRig', [x, y, rz])]
```

### jel2 group

```python
jel_lst = []
x = 'rigid'
y = 'rigid'
rz = ['lin', 0.1]
jel_lst.append(utils.param2adap('gjelPin', [x, y, rz]))
x = ['lin', 0.1]
y = 'rigid'
rz = ['lin', 0.1]
jel_lst.append(utils.param2adap('gjelRol', [x, y, rz]))
```

### cbm2 group

```python
cbm_lst=[]
for i in sections_lst:
    for s in i:
        sectID=s.ID
        try:
            matID=s.conf_mat
            sectArea=s.area
        except:
            matID=s.mat
            sectArea=s.area
        for m in material_lst:
            if m.ID == matID: 
                density = m.density
                break
        cbm_lst.append(['g'+sectID+'m',sectArea*density]) # units of kg/mm
```

### group_lst

Group all section groups as `group_lst`.

```python
group_lst = [cbp_lst,lnk_lst,jel_lst,cbm_lst]
```

## Structural nodes

### Beam nodes

```python
nodes=pd.DataFrame([[i+1,i*250,0,'#','beam node '+str(i+1)] for i in range(25)],columns= ['name','x','y','#','ID'])
nodes
```

### Restrained nodes

Add beam nodes at the bottom of the section, which will be connected to the cbp elements with rigid links. The from the centroid to the bottom fibre is -428mm.

```python
nodes_temp = pd.DataFrame([[26,0,-beam_section1.centr_top,'#','beam support node 1'],
                           [27,6000,-beam_section1.centr_top,'#','beam support node 2'],
                           [28,0,-beam_section1.centr_top,'#','support node 1'],
                           [29,6000,-beam_section1.centr_top,'#','support node 2']],
                          columns= ['name','x','y','#','ID'])
nodes = pd.concat([nodes,nodes_temp],ignore_index = True)
nodes
```

## Non-structural nodes

```python
non_st_nodes=pd.DataFrame([['nst1',999999,0,'#','beam centroid'],
                          ['nst2',999999,-beam_section1.centr_top,'#','beam bottom']],columns= ['name','x','y','#','ID'])
non_st_nodes
```

## Restraints

```python
restraints = pd.DataFrame([[28,'x+y+rz','#','restraint 1'],[29,'x+y+rz','#','restraint 2']],columns=['name','direction','#','ID'])
restraints
```

## Element connectivity

### cbp2 elements

The connectivity of cbp elements is reversed so that the beam has an inversed T-section.

```python
elm_cbp = [[startCBP+i,'gB1',nodes['name'][i+1],nodes.loc[i]['name'],'#',"beam cbp "+str(i+1) ] for i in range(24)]
elm_cbp = pd.DataFrame(elm_cbp,columns = ['name', 'section', 'node 1', 'node 2', '#', 'ID'])
elm_cbp
```

### jel2 elements

```python
elm_jel=pd.DataFrame([[startJEL,'gjelPin',26,28,'nst2','#','jel support 1'],
                     [startJEL+1,'gjelRol',27,29,'nst2','#','jel support 2']],
                         columns=['name','section','node 1','node 2','nst node','#','ID'])
elm_jel
```

### lnk2 elements

```python
elm_lnk=pd.DataFrame([[startLNK,'glnkRig',1,26,'#','link support 1'],
                     [startLNK+1,'glnkRig',25,27,'#','link support 2']],
                         columns=['name','section','node 1','node 2','#','ID'])
elm_lnk
```

### cbm elements

```python
starID=startCBM 
elm_cbm=elm_cbp.copy()
elm_cbm.at[:,'section']=elm_cbm['section']+'m'
elm_cbm.at[:,'ID']='cbm '+elm_cbm['ID']
elm_cbm.at[:,'name']=[starID+i for i in range(len(elm_cbm))]
elm_cbm
```

## Print lines in .dat files

```python
# Init all_lines
all_lines = np.array("# Init")

# Write file info
line = utils.str_joint(["#",project_name,str(day.today())])
all_lines = add(line)
all_lines = add("analysis 2d dynamics")
```

```python
all_lines = add('#')
all_lines = add('materials')
all_lines = add('mat.name model properties')
for i in material_lst:
    all_lines = add(i.adaptic_print())
    
all_lines = add('#')
all_lines = add('sections')
if section_rcts_lst !=[]:
    all_lines = add('type = rcts')
    all_lines = add('sec.name mat.name dimensions')
    for i in section_rcts_lst:
        all_lines = add(i.adaptic_print())
        
all_lines = add('#')
all_lines = add('groups')
if not elm_cbp.empty:
    all_lines = add('type.of.element = cbp2')
    all_lines = add('grp.name sec.name monitoring.points')
    for i in cbp_lst:
        all_lines = add(utils.str_joint(i))
if not elm_cbm.empty:
    all_lines = add('type.of.element = cbm2')
    all_lines = add('grp.name mass')
    for i in cbm_lst:
        all_lines = add(utils.str_joint(i))
if not elm_lnk.empty:
    all_lines = add('type.of.element = lnk2')
    all_lines = add('grp.name stiffness.parameters')
    for i in lnk_lst:
        all_lines = add(i.printout)
if not elm_jel.empty:
    all_lines = add('type.of.element = jel2')
    all_lines = add('grp.name curve.types parameters')
    for i in jel_lst:
        all_lines = add(i.printout)

all_lines = add('#')
all_lines = add('structural.nodal.coordinates')
all_lines = add('nod.n x y')
for i in range(len(nodes)):
    all_lines = add(nodes.loc[[i]].to_string(header=False, index=False))

all_lines = add('#')
all_lines = add('non.structural.nodal.coordinates')
all_lines = add('nod.n x y')
for i in range(len(non_st_nodes)):
    all_lines = add(non_st_nodes.loc[[i]].to_string(header=False, index=False))
    
all_lines = add('#')
all_lines = add('element.connectivity')
all_lines = add('elm.name grp.name nod.')
all_lines = add('# name node1, node2, (node3), start, end, length, ID')
if not elm_cbp.empty:
    all_lines = add('# cbp elements')
    for i in range(len(elm_cbp)):
        all_lines = add(elm_cbp.loc[[i]].to_string(header=False, index=False))
if not elm_jel.empty:
    all_lines = add('# jel elements')
    for i in range(len(elm_jel)):
        all_lines = add(elm_jel.loc[[i]].to_string(header=False, index=False))
if not elm_lnk.empty:
    all_lines = add('# lnk elements')
    for i in range(len(elm_lnk)):
        all_lines = add(elm_lnk.loc[[i]].to_string(header=False, index=False))
if not elm_cbm.empty:
    all_lines = add('# cbm elements')
    for i in range(len(elm_cbm)):
        all_lines = add(elm_cbm.loc[[i]].to_string(header=False, index=False))

all_lines = add('#')
all_lines = add('restraints')
all_lines = add('nod.n direction')
for i in range(len(restraints)):
    all_lines = add(restraints.loc[[i]].to_string(header=False, index=False))
```

## analysis parameters

```python
# integration
integration = [['#'],['integration'],
['scheme = hilber'],
['alpha = -0.3'],
['beta = 1.21'],
['gamma = 0.8']]
analysis = integration
# equilibrium stages
stages=[['#'],['equilibrium.stages'],
['end.of.stage steps'],
[500, 1000]]
analysis.extend(stages)
```

### solution

```python
solution = [['solution.of.equations'],
           ['solver = mumps'],
           ['#']]
analysis.extend(solution)
```

### iterative strategy

```python
lin_search_bool = False
```

```python
if lin_search_bool: 
    iters=8
    stepRed=50
else: 
    iters=20
    stepRed=10
iterative_str = [['iterative.strategy'],
['number.of.iterations = ',iters],
['initial.reformations = ',iters, '#if equal to number.of.iterations it is Full N-R'],
['step.reduction = ',stepRed, '# step reduction'],
['divergence.iteration = ',iters, '# no of iteration after which divergence is checked'],
['maximum.convergence = ',1.00E+20, '# iterations terminated if this convergence is exceeded'],
['#']]
analysis.extend(iterative_str)
```

### linear search

```python
if lin_search_bool:
    lin_search=[
        ['lin.search'],
        ['beta = 0.5'],
        ['emin = 0.2'],
        ['emax = 2'],
        ['trials = 5'],
        ['#']]
    analysis.extend(lin_search)
```

### convergence criteria

```python
convergance = [['convergence.criteria'],
['tolerance = ',1.00E-05, '# tolerance against which the convergence norm is compared'],
['work.ref = ',1.60E+09,'#work reference'],
['# force.ref = 1.00E+07 # used with the norm of out-of-balance forces to determine convergence'],
['# moment.ref = 1.00E+05 # used with the norm of out-of-balance moments to determine convergence'],
['#']]
analysis.extend(convergance)
```

### output

```python
freq = 1 # results every `freq` output
output = [['output'],
['frequency '+str(freq)+' stress # output results for all equilibrium step, including step reductions'],
['#'],
['end']]
analysis.extend(output)
```

### print all the analysis parameters

```python
for i in analysis:
    all_lines = add(utils.str_joint(i))
```

## save file

Convert `all_lines` variable to a _*.dat_ file.

```python
if save:
    f= open(project_name+".dat","w+")
    for i in all_lines:
         f.write(i[0]+"\n")
    f.close()
    print("File {0}.dat has been succesfully created.".format(project_name))
```

```python
if save:
    import pickle
    model_pkl = [material_lst,sections_lst,group_lst,nodes,non_st_nodes,restraints,elm_cbp,elm_jel,elm_lnk,elm_cbm]
    with open(project_name+'.pkl', "wb") as f:
        pickle.dump(model_pkl, f)
    f.close()    
    print('Saved pickle as '+project_name+'.pkl')
```

## Plot the structure

```python
def plot_labels(label,color,linestyle='-',linewidth=3,marker='',markersize=3):
    if not label in labels: 
        lines.append(Line2D([0], [0], color=color, linewidth=linewidth, linestyle=linestyle,marker=marker,markersize=markersize))
        labels.append(label)
    return labels
```

```python
fig, ax = plt.subplots(figsize=(12,3))
ax.grid(which='major', linestyle=':', linewidth='0.5', color='black')
markersize=3
lines, labels, colors = [], [], []
for i in range(len(elm_cbp)):
    x=[float(utils.df_value(nodes,elm_cbp['node 1'][i],'name','x')),float(utils.df_value(nodes,elm_cbp['node 2'][i],'name','x'))]
    y=[float(utils.df_value(nodes,elm_cbp['node 1'][i],'name','y')),float(utils.df_value(nodes,elm_cbp['node 2'][i],'name','y'))]
    ax.plot(x,y,'-', linewidth=2, marker='o',markersize=markersize,color = 'blue')
    labels=plot_labels('cbp','blue')            
for i in range(len(elm_lnk)):
    x=[float(utils.df_value(nodes,elm_lnk['node 1'][i],'name','x')),float(utils.df_value(nodes,elm_lnk['node 2'][i],'name','x'))]
    y=[float(utils.df_value(nodes,elm_lnk['node 1'][i],'name','y')),float(utils.df_value(nodes,elm_lnk['node 2'][i],'name','y'))]
    ax.plot(x,y,'-', linewidth=2, marker='s', markersize=markersize,color = 'green')
    labels=plot_labels('rigid link','green')       
for i in range(len(elm_jel)):
    x=[float(utils.df_value(nodes,elm_jel['node 1'][i],'name','x')),float(utils.df_value(nodes,elm_jel['node 2'][i],'name','x'))]
    y=[float(utils.df_value(nodes,elm_jel['node 1'][i],'name','y')),float(utils.df_value(nodes,elm_jel['node 2'][i],'name','y'))]
    if elm_jel['section'][i]=="gjelPin":
        ax.plot(x,y,'-', marker='^',markersize=10, color = 'k')
        labels=plot_labels('pin','k',linestyle='',marker='^',markersize=10)
    elif elm_jel['section'][i]=="gjelRol":
        ax.plot(x,y,marker='X',markersize=10, color = 'k')
        labels=plot_labels('roller','k',linestyle='',marker='X',markersize=10)
for i in range(len(load)):
    x=[float(utils.df_value(nodes,load['node'][i],'name','x'))]
    y=[float(utils.df_value(nodes,load['node'][i],'name','y'))]
    ax.plot(x,y,'', marker='v',markersize=10, color = 'r')
    labels=plot_labels('load','r',linestyle='',marker='v',markersize=10)

legend = ax.legend(lines,labels, bbox_to_anchor=(1, 1.05))

plt.axis('equal')
ax.set_title('simply supported beam') # project_name
plt.show()
```
